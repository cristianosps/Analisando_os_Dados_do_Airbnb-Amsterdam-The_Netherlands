### Transformaes necessrias para a anlise

Aps visualizada a distribuio das variveis *price* e *minimum_nights* atravs dos diagramas de Boxplot e Histograma, ser realizada a anlise dos valores e as transformaes necessrias sero efetuadas no conjunto de dados utilizado para melhor organizao/ entendimento.

.


Primeiramente, ser feita uma cpia do conjunto de dados original e a eliminao da coluna *neighbourhood_group*, pois no h valores vlidos que possam ser utilizados durante a anlise. 

```
# Primeiramente, ser realizada uma cpia do DataFrame original para outro, onde estaro os dados aps transformaes.

df_clean = df.copy()

# A coluna "neighbourhood_group" no ser necessria, pois no se encontra preenchida com valores vlidos ou que possam auxiliar na anlise.

df_clean.drop("neighbourhood_group", axis=1, inplace=True)
```

Com base na anlise dos diagramas,  clara a existncia de outliers que devemos tratar utilizando tcnicas para identificar os pontos de corte para eliminao dos outliers, assim diminuindo a distoro entre os valores do conjunto de dados para essas variveis.

", width="50%"></p>


Tomando como base os componentes do Boxplot, o tratamento para eliminao dos outliers a ser realizado seguir os seguintes passos:

* Definir o Q1 e Q3 para as variveis "Price" e "Minimum_nights";
* Calcular o Intervalo Interquartil (IQR - Interquartile range) para as variveis;
* Definir o limite superior e inferior para remoo dos outliers;
* Remover os outliers.

### Identificao dos valores de corte dos outliers para a varivel "price"


```
# Identificao dos valores de corte dos outliers para a varivel price

q1_price = df_clean.price.quantile(.25)
q3_price = df_clean.price.quantile(.75)
IQR_price = q3_price - q1_price
print('Valor IQR para varivel price: ', IQR_price)

# Definir os Limites Superior e Inferior                                      

inf_price = q1_price - 1.5 * IQR_price
sup_price = q3_price + 1.5 * IQR_price

print('\nValor limite inferior para varivel price: ', inf_price)
print('Valor limite superior para varivel price: ', sup_price)
```

```
# Quantidade de linhas as serem excludas com base nos valores de corte

print("\nTotal de entradas do conjunto de dados:\t {}".format(df_clean.shape[0]))
print("Quantidade de entradas acima de 306.0:\t {}".format(len(df_clean[df_clean.price > sup_price])))
print("Percentual de entradas acima de 306.0:\t {:.2f}%".format((len(df_clean[df_clean.price > sup_price]) / df_clean.shape[0])*100))
```

### Identificao dos valores de corte dos outliers para a varivel "minimum_nights"

```
# Identificao dos valores de corte dos outliers para a varivel minimum_nights

q1_minimum_nights = df_clean.minimum_nights.quantile(.25)
q3_minimum_nights = df_clean.minimum_nights.quantile(.75)
IQR_minimum_nights = q3_minimum_nights - q1_minimum_nights
print('Valor IQR para varivel minimum_nights: ', IQR_minimum_nights)

# Definir os Limites Superior e Inferior                                      

inf_minimum_nights = q1_minimum_nights - 1.5 * IQR_minimum_nights
sup_minimum_nights = q3_minimum_nights + 1.5 * IQR_minimum_nights

print('\nValor limite inferior para varivel minimum_nights: ', inf_minimum_nights)
print('Valor limite superior para varivel minimum_nights: ', sup_minimum_nights)
```

```
# Quantidade de linhas as serem excludas com base nos valores de corte

print("\nTotal de entradas do conjunto de dados:\t {}".format(df_clean.shape[0]))
print("Quantidade de entradas acima de 4.5:\t {}".format(len(df_clean[df_clean.minimum_nights > sup_minimum_nights])))
print("Percentual de entradas acima de 4.5:\t {:.2f}%".format((len(df_clean[df_clean.minimum_nights > sup_minimum_nights]) / df_clean.shape[0])*100))
```

### Remoo dos outliers do conjunto de dados para as variveis "price" e "minimum_nights"

```
# Remoo dos outliers do conjunto de dados para as variveis Price e Minimum_nights

df_clean.drop(df_clean[df_clean.price > sup_price].index, axis=0, inplace=True)
df_clean.drop(df_clean[df_clean.price == 0.0].index, axis=0, inplace=True)
df_clean.drop(df_clean[df_clean.minimum_nights > sup_minimum_nights].index, axis=0, inplace=True)

print('Dimenses originais do conjunto de dados: ', df.shape)
print('Novas dimenses do cojunto de dados\t: ',df_clean.shape)

print("\nPS: As novas dimenses foram atribuidas aps remoo da varivel neighbourhood_group e dos valores outliers para 'price' e 'minimum_nights'")
```
