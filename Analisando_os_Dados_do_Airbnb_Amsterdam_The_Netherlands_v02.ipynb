### Transformações necessárias para a análise

Após visualizada a distribuição das variáveis *price* e *minimum_nights* através dos diagramas de Boxplot e Histograma, será realizada a análise dos valores e as transformações necessárias serão efetuadas no conjunto de dados utilizado para melhor organização/ entendimento.

.


Primeiramente, será feita uma cópia do conjunto de dados original e a eliminação da coluna *neighbourhood_group*, pois não há valores válidos que possam ser utilizados durante a análise. 

```
# Primeiramente, será realizada uma cópia do DataFrame original para outro, onde estarão os dados após transformações.

df_clean = df.copy()

# A coluna "neighbourhood_group" não será necessária, pois não se encontra preenchida com valores válidos ou que possam auxiliar na análise.

df_clean.drop("neighbourhood_group", axis=1, inplace=True)
```

Com base na análise dos diagramas, é clara a existência de outliers que devemos tratar utilizando técnicas para identificar os pontos de corte para eliminação dos outliers, assim diminuindo a distorção entre os valores do conjunto de dados para essas variáveis.

", width="50%"></p>


Tomando como base os componentes do Boxplot, o tratamento para eliminação dos outliers a ser realizado seguirá os seguintes passos:

* Definir o Q1 e Q3 para as variáveis "Price" e "Minimum_nights";
* Calcular o Intervalo Interquartil (IQR - Interquartile range) para as variáveis;
* Definir o limite superior e inferior para remoção dos outliers;
* Remover os outliers.

### Identificação dos valores de corte dos outliers para a variável "price"


```
# Identificação dos valores de corte dos outliers para a variável price

q1_price = df_clean.price.quantile(.25)
q3_price = df_clean.price.quantile(.75)
IQR_price = q3_price - q1_price
print('Valor IQR para variável price: ', IQR_price)

# Definir os Limites Superior e Inferior                                      

inf_price = q1_price - 1.5 * IQR_price
sup_price = q3_price + 1.5 * IQR_price

print('\nValor limite inferior para variável price: ', inf_price)
print('Valor limite superior para variável price: ', sup_price)
```

```
# Quantidade de linhas as serem excluídas com base nos valores de corte

print("\nTotal de entradas do conjunto de dados:\t {}".format(df_clean.shape[0]))
print("Quantidade de entradas acima de 306.0:\t {}".format(len(df_clean[df_clean.price > sup_price])))
print("Percentual de entradas acima de 306.0:\t {:.2f}%".format((len(df_clean[df_clean.price > sup_price]) / df_clean.shape[0])*100))
```

### Identificação dos valores de corte dos outliers para a variável "minimum_nights"

```
# Identificação dos valores de corte dos outliers para a variável minimum_nights

q1_minimum_nights = df_clean.minimum_nights.quantile(.25)
q3_minimum_nights = df_clean.minimum_nights.quantile(.75)
IQR_minimum_nights = q3_minimum_nights - q1_minimum_nights
print('Valor IQR para variável minimum_nights: ', IQR_minimum_nights)

# Definir os Limites Superior e Inferior                                      

inf_minimum_nights = q1_minimum_nights - 1.5 * IQR_minimum_nights
sup_minimum_nights = q3_minimum_nights + 1.5 * IQR_minimum_nights

print('\nValor limite inferior para variável minimum_nights: ', inf_minimum_nights)
print('Valor limite superior para variável minimum_nights: ', sup_minimum_nights)
```

```
# Quantidade de linhas as serem excluídas com base nos valores de corte

print("\nTotal de entradas do conjunto de dados:\t {}".format(df_clean.shape[0]))
print("Quantidade de entradas acima de 4.5:\t {}".format(len(df_clean[df_clean.minimum_nights > sup_minimum_nights])))
print("Percentual de entradas acima de 4.5:\t {:.2f}%".format((len(df_clean[df_clean.minimum_nights > sup_minimum_nights]) / df_clean.shape[0])*100))
```

### Remoção dos outliers do conjunto de dados para as variáveis "price" e "minimum_nights"

```
# Remoção dos outliers do conjunto de dados para as variáveis Price e Minimum_nights

df_clean.drop(df_clean[df_clean.price > sup_price].index, axis=0, inplace=True)
df_clean.drop(df_clean[df_clean.price == 0.0].index, axis=0, inplace=True)
df_clean.drop(df_clean[df_clean.minimum_nights > sup_minimum_nights].index, axis=0, inplace=True)

print('Dimensões originais do conjunto de dados: ', df.shape)
print('Novas dimensões do cojunto de dados\t: ',df_clean.shape)

print("\nPS: As novas dimensões foram atribuidas após remoção da variável neighbourhood_group e dos valores outliers para 'price' e 'minimum_nights'")
```
